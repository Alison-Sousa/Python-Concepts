No exempo a seguir o método arange da biblioteca Numpy é utilizado para gerar conjuntos de dados.

sintaxe: variavel =  np.arange(valor_inicial, valor_final, valor_incremento)
exemplo:

conjunto_1 =  np.arange(0, 99, 1), nesse exemplo a variável conjunto receberá uma lista de valores compreendida entre 0 e 99 (não incluso) com incremento 1, isso é, variando de 1 em 1.

conjunto_2 =  np.arange(0, 99, 3), nesse exemplo a variável conjunto receberá uma lista de valores compreendida entre 0 e 99 (não incluso) com incremento 3, isso é, variando de 3 em 3.

```python
import numpy as np
conjunto_1 = np.arange(0,99,1) 
conjunto_2 = np.arange(0,99,3) 
conjunto_3 = np.arange(7, 33, 2)
print("Conjunto 1 - incremento de 1 em 1")
print(conjunto_1)
print(" Conjunto 2 - incremento de 3 em 3 ")
print (conjunto_2)
print("Conjunto 3 - conjunto de valores entre 7 e 33 com incremento de 2 em 2")
print(conjunto_3)
```

Alterar o exemplo para criar um conjunto de valores entre 1000 e 2000 com incremento de 200 em 200 e mostre o conjunto gerado.

a biblioteca Numpy disponibiliza várias funções para manipulação de dados  numéricos, dentre eles temos:

max() - retorna o maior valor de um conjunto

min() - retorno o menor valor de um conjunto

sum() - retorna a soma entre todos os valores do conjunto

mean() - retorna a média entre todos os valores do conjunto


A sintaxe para utilização das funções é:



> ***nome_conjunto.nome_funcao()***

Nas linhas 4 a 7 do exemplo a seguir é possível visualizar exemplos de utilização.



```python
import numpy as np
conjunto = np.arange(158,300,2)
print(conjunto)
print("Elemento de maior valor ...", conjunto.max())
print("Elemento de menor valor ...", conjunto.min())
print("Soma dos valores ...", conjunto.sum())
print("Média dos valores ...", conjunto.mean())
```

import matplotlib.pyplot as plt

A plotagem do gráfico de linhas pode ser feita conforme a sintaxe:

plt.plot(eixo x, eixo y)

No exemplo a seguir, na linha 5, ilustra-se a instrução para plotagem do gráfico de alunos e idades, na linha 6 a instrução plt.show() apresenta o gráfico.

```python
import numpy as np
import matplotlib.pyplot as plt
alunos = ['Sandra', 'Pedro', 'Lucas', 'Gabriela','Ricardo', 'Matheus', 'Laurinha', ' Victor']
idade = [51, 19, 16, 21, 27, 19, 19, 19]
plt.plot(alunos, idade)
plt.show()
```

# **Gráfico de Barras**

```python
import numpy as np
import matplotlib.pyplot as plt
alunos = ['Sandra', 'Pedro', 'Lucas', 'Gabriela','Ricardo', 'Matheus', 'Laurinha', ' Victor']
idade = [51, 19, 16, 21, 27, 19, 19, 19]
width_n = 0.5
bar_color = 'k' 
plt.bar(alunos, idade, width=width_n, color=bar_color)
plt.show()
```
