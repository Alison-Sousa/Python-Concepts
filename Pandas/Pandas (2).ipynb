# **Ordenação**

Pode-se ordenar o conjunto de dados por uma determinada coluna, no exemplo a seguir vamos ordenar por disciplina.

> variavel.sort_values(by="nome da coluna")

* sort_values: instrução papra ordenar o conjunto de dados
* by="nome da coluna": indicar a coluna pela qual o conjunto de dados será ordenado


Exemplo

> dados.sort_values(by="Disciplina")

Vamos verificar quais são os alunos que obtiveram média maior ou igual a 9,0 em alguma disciplina e atribuir o resultado a uma variável, em seguida o conjunto de dados será ordenado das maiores médias para as menores

```python
import pandas as pd
dados = pd.read_excel("/content/drive/MyDrive/BDados/Notas/Notas.xlsx")
dados.Média
x = dados[dados.Média >= 9]
x.sort_values(by="Média", ascending=False)
```

# **Seleção de dados - Filtro**

Pode-se selecionar dados de acordo com condições combinadas, no exemplo a seguir serão selecionadas as linhas que possuem Média >= 9 e Disciplina == 'Biologia'

**Lembre-se que: **
* operador e (and)  é representado por **&**
* == é o símbolo utilizado para comparação de igualdade
* = é utilizado para atribuição

Sintaxe:

variável[operação para seleção de dados]

onde 

* variável: nome da variável que recebeu o arquivo (dataSet)
* operação para seleção de dados: é possível utilizar expressões relacionais usando os operadores >, >=, <, <=, ==, !=  e combinar mais de uma utilizando os operadores  & (e) ou | (ou)


```python
import pandas as pd
dados = pd.read_excel("/content/drive/MyDrive/BDados/Notas/Notas.xlsx")
dados.Média
dados[(dados.Média >= 9) & (dados.Disciplina == 'Biologia')]
```

Pode-se selecionar dados de acordo com condições combinadas, no exemplo a seguir serão selecionadas as linhas que possuem Média >= 9 ou Disciplina == 'Biologia', o resultado é armazenado na variável **resultado**.
O operador ou é representado por **|**

```python
import pandas as pd
dados = pd.read_excel("/content/drive/MyDrive/BDados/Notas/Notas.xlsx")
dados.Média
resultado = dados[(dados.Média >= 9) | (dados.Disciplina == 'Biologia')]
resultado.sort_values(by="Disciplina")
```

# **Seleção de colunas**

É possível selecionar uma ou mais colunas de um dataSet criar um subconjunto de dados. Esse novo conjunto de dados pode ser manipulado diretamente ou armazenados em arquivos csv, excel ou bases de dados.

No exemplo a seguir, na linha 3 são selecionadas as colunas* Aluno, Disciplina *e* *Média*, essas colunas fazem parte do dataSet *dados*. Os nomes das colunas (campos) devem ser iguais aos nomes originais. Em seguida, as linhas são filtradas e armazenadas em um novo dataSet.

Sintaxe para selecionar as colunas:

> variável_colunas = ['nome da coluna', 'nome da coluna', etc.] 

Exemplo

> colunas_selecionadas = ['Aluno', 'Disciplina', 'Média']


Sintaxe para gerar o subconjunto

> variável_dados = dataSet.filter(items=variável_colunas)

Exemplo 
> dados_selecionados = dados.filter(items=colunas_selecionadas)



```python
import pandas as pd
dados = pd.read_excel("/content/drive/MyDrive/BDados/Notas/Notas.xlsx")
colunas_selecionadas = ['Aluno', 'Disciplina', 'Média']
dados_selecionados = dados.filter(items=colunas_selecionadas)
dados_selecionados.head()
```

# **Gravação de dados**

Os dados selecionados de arquivos ou bancos de dados podem ser filtrados, tratados e podem ser armazenados (gravados) em arquivos ou bancos de dados.

# **Agrupamento de dados**

Os dados podem ser agrupados por campos possibilitando a sumarização de resultados, por exemplo apresentando o número de registros, a média de valores, a identificação do maior e/ou do menor valor dentre outros.

Sintaxe
> variável_dados.groupby('coluna').função()

onde

* variável_dados: variável que armazena os dados selecionados
* groupby: instrução que promove o agrupamento de dados
* coluna: representa a coluna ou lista de colunas que serão utilizadas para o agrupamento dos dados. O nome da coluna deve ser idêntico ao nome do dataSet.
* função(): função que será utilizada para sumarizar os dados

 **Algumas funções são:**

         * count( ) - retorna a quantidade de registros
         * mean( )  - retorna a média entre os valores
         * sum ( ) - retorna a soma dos valores

**Exemplo**

> dados_selecionados.groupby('Disciplina').count()

```python
import pandas as pd
dados = pd.read_excel("/content/drive/MyDrive/BDados/Notas/Notas.xlsx")
colunas_selecionadas = ['Aluno', 'Disciplina']
dados_selecionados = dados.filter(items=colunas_selecionadas)
dados_selecionados.groupby('Disciplina').count()
```

No exemplo a seguir, calcula-se a pontuação do estudante com base na somatória das suas médias.

```python
import pandas as pd
dados = pd.read_excel("/content/drive/MyDrive/BDados/Notas/Notas.xlsx")
colunas_selecionadas = ['Aluno', 'Disciplina', 'Média']
dados_selecionados = dados.filter(items=colunas_selecionadas)
dados_selecionados.groupby('Aluno').sum()
```
