**Gráfico a seguir representa o crescimento do PIB até 2026, comparando com 2025, dos países que também possuem armas nucleares, de acordo com (IFM,2022).**

**Fonte**: https://www.imf.org/external/datamapper/NGDP_RPCH@WEO/RUS/USA/GBR/CHN/FRA/IND/PAK/ISR/PRK

```python
import pandas as pd #manipulação e ordenamento dos dados
import matplotlib.pyplot as plt # biblioteca para plotagem de gráficos
PIB = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/imf-dm-export-20220508.csv', sep = ',') # selecionando a base de dados csv.

colunas_selecionadas = ['Real GDP growth (Annual percent change)', '2025', '2026'] # seleção das colunas que será exposto no gráfico.
dados_selecionados = PIB.filter(items=colunas_selecionadas) # filtro das colunas que será exposto no gráfico.

dados_selecionados.plot('Real GDP growth (Annual percent change)') # filtro da coluna que será exposto no gráfico.

x= plt.xlabel('Países', fontsize=16) # Se ajusta o eixo e a fonte.
y = plt.ylabel('PIB%', fontsize=16) # Se ajusta o eixo  e a fonte.
plt.title('Crescimento % PIB dos países com armas nucleares', fontsize=16) # Se ajusta o título e a fonte.
plt.rcParams[('figure.figsize')] =(30,25) # Se ajusta o tamanho do gráfico para ficar legível.

plt.show() # Mostrar o gráfico e o resultado final da programação.

```

**O gráfico a seguir mostra a quantidade de testes nucleares que a Rússia realizou entre (1938-2022),de acordo com (Our World in Data, 2022).**

**Fonte:** https://ourworldindata.org/nuclear-weapons

```python
import pandas as pd #manipulação e ordenamento dos dados
import matplotlib.pyplot as plt  # biblioteca para plotagem de gráficos

nuclear = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/number-of-nuclear-weapons-tests.csv',sep = ',') # selecionando a base de dados csv.


colunas_selecao = ['Entity', 'nuclear_weapons_tests'] # seleção das colunas que será exposto no gráfico.
selecao = nuclear.filter(items=colunas_selecao) # filtro das colunas que será exposto no gráfico.
plot2= selecao[(selecao.Entity == "Russia")] # filtro da coluna e variável escolhida que será exposto no gráfico.

plot2.plot('Entity') # plotando o gráfico no Python.

plt.xlabel('Country', fontsize=16) # Se ajusta o eixo e a fonte.
plt.ylabel('Number Of Nuclear Weapons', fontsize=16) # Se ajusta o eixo e a fonte.
plt.title('Number Of Nuclear Weapons Tests (1938 - 2022)', fontsize=16) # Se ajusta o título e a fonte.

plt.show() # Mostrar o gráfico e o resultado final da programação.
```

**O gráfico a seguir mostra a posição(status) nuclear da Rússia entre (1938-2022),de acordo com (Our World in Data, 2022).**

**Fonte:** https://ourworldindata.org/nuclear-weapons

```python
import pandas as pd #manipulação e ordenamento dos dados
import matplotlib.pyplot as plt  # biblioteca para plotagem de gráficos

paises_nuclear = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/country-position-nuclear-weapons.csv') # selecionando a base de dados csv.

position_nuclear = ['Entity', 'nuclear_weapons_status'] # seleção das colunas que será exposto no gráfico.
selection = paises_nuclear.filter(items=position_nuclear) # filtro das colunas que será exposto no gráfico.
graph = selection[(selection.Entity == "Russia")] # filtro da coluna e variável escolhida que será exposto no gráfico.

graph.plot('Entity')  # plotando o gráfico no Python.

plt.xlabel('Country', fontsize=16)  # Se ajusta o eixo e a fonte.
plt.ylabel('Nuclear Weapons Status', fontsize=16)  # Se ajusta o eixo e a fonte.
plt.title('Nuclear Weapons Status (1938 - 2022)', fontsize=16) # Se ajusta o título e a fonte.

plt.show() # Mostrar o gráfico e o resultado final da programação.
```

```python
import pandas as pd #manipulação e ordenamento dos dados
import matplotlib.pyplot as plt  # biblioteca para plotagem de gráficos

armas = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Sheet 1_data.csv',sep = ',') # selecionando a base de dados csv.
colunas_focadas = ['Country', 'Total Inventory'] # seleção das colunas que será exposto no gráfico.
selected = armas.filter(items=colunas_focadas) # filtro das colunas que será exposto no gráfico.

selected.plot('Country')  # plotando o gráfico no Python.

plt.xlabel('País', fontsize=16)  # Se ajusta o eixo e a fonte.
plt.ylabel('Quantidade do Arsenal', fontsize=16)  # Se ajusta o eixo e a fonte.
plt.title('Arsenal Nuclear em 2022', fontsize=16) # Se ajusta o título e a fonte.

plt.show() # Mostrar o gráfico e o resultado final da programação.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import investpy

results = investpy.search_quotes(text = 'SBER', products= ['stocks'], countries= ['russia'], n_results = 50) # Delimitando o tipo de produto que procuro, que é a ação diária da Sberbank.

for results in results[1:2]: #Focando nos resultados iniciais.
  print(results)

jpm = results.retrieve_historical_data(from_date = '01/01/2018', to_date = '13/05/2022') # Selecionando o período das cotações.

jpm.head() # Lendo e "printando" as variáveis trabalhadas.

plt.xlabel('Data', fontsize=16)
plt.ylabel('Preço', fontsize=16)
plt.title('Açoes Sberbank antes e durante a invasão da Rússia na Ucrânia em 2022', fontsize=16)


jpm['Close'].plot()
```
