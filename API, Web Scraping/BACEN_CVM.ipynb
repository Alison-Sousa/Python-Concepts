```python
import requests
import xml.etree.ElementTree as ET

# URL do feed XML
url = "https://www.bcb.gov.br/api/feed/app/normativos/normativos?ano=2021"

# Fazer uma solicitação GET para o URL do feed
response = requests.get(url)

# Analisar o feed XML usando a biblioteca ElementTree
root = ET.fromstring(response.content)

# Inicializar uma lista vazia de entradas
entries = []

# Iterar sobre todas as entradas no feed
for entry in root.findall('{http://www.w3.org/2005/Atom}entry'):
    # Extrair o título, data de atualização e conteúdo da entrada
    title = entry.find('{http://www.w3.org/2005/Atom}title').text
    updated = entry.find('{http://www.w3.org/2005/Atom}updated').text
    content = entry.find('{http://www.w3.org/2005/Atom}content').text

    # Adicionar uma tupla (title, updated, content) à lista de entradas
    entries.append((title, updated, content))

# Ordenar a lista de entradas com base na data de atualização (do mais recente para o mais antigo)
entries = sorted(entries, key=lambda x: x[1], reverse=True)

# Imprimir as três entradas mais recentes
for i, entry in enumerate(entries[:3]):
    print("Entrada", i+1)
    print("Título:", entry[0])
    print("Data de atualização:", entry[1])
    print("Conteúdo:", entry[2])
    print()
```
